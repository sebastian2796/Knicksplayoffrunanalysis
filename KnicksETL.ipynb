```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('knicks2026playoffrun.csv')
df2 = pd.read_csv('knicks2025playoffrun.csv')
```

```python
df1["Season"] = 2026
df2["Season"] = 2025

combined_df = pd.concat(
    [df1, df2],
    ignore_index=True
)
```

```python
combined_df = pd.concat(
    [df1, df2],
    ignore_index=True
)
```

```python
combined_df = combined_df[
    combined_df["Rk"] != "Rk"
]
```

```python
numeric_cols = [
    "Tm",
    "Opp",
    "ORtg",
    "DRtg",
    "Pace",
    "TS%",
    "eFG%",
    "TOV%"
]

for col in numeric_cols:
    combined_df[col] = pd.to_numeric(
        combined_df[col],
        errors="coerce"
    )
```

```python
print(combined_df.columns)
```

```python
combined_df["Net_Rating"] = (
    combined_df["ORtg"] -
    combined_df["DRtg"]
)
```

```python
combined_df["Tm"] = pd.to_numeric(
    combined_df["Tm"],
    errors="coerce"
)

combined_df["Opp.1"] = pd.to_numeric(
    combined_df["Opp.1"],
    errors="coerce"
)

combined_df["Point_Diff"] = (
    combined_df["Tm"] -
    combined_df["Opp.1"]
)

combined_df["Win"] = np.where(
    combined_df["Point_Diff"] > 0,
    1,
    0
)

combined_df["Close_Game"] = np.where(
    combined_df["Point_Diff"].abs() <= 5,
    1,
    0
)
```

```python
combined_df = combined_df.rename(
    columns={"Opp.1": "Opponent_PTS"}
)
```

```python
combined_df.to_excel(
    "Knicks_Playoff_Analysis.xlsx",
    index=False
)
```
